# Code for bar plot of Statewise total reported rape victims from 2001-2010
# (Graph 1)

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from pandas.tools.plotting import scatter_matrix
import seaborn as sns
%matplotlib inline
# victims of rape DataFrame
rape_victims = pd.read_csv('20_Victims_of_rape.csv')
rape_victims_by_state = rape_victims.groupby('Area_Name').sum()   #grouping values based on area and taking their sum
c = rape_victims_by_state.drop('Year', axis = 1, inplace = True)
plt.subplots(figsize = (11, 6))
sorted_rape_by_state = rape_victims_by_state['Rape_Cases_Reported'].sort_values(ascending = False)
ax = sorted_rape_by_state.plot.bar()
ax.set_xlabel('Area Name')
ax.set_ylabel('Total Number of Reported Rape Victims from 2001 to 2010')
ax.set_title('Statewise total Reported Rape Victims throught the Years 2001 to 2010')
plt.show()
```

# Code for plotting Number of police officers vs Number of rapes in Madhya Pradesh (Graph 2)

```python
import csv
total_rapes = pd.read_csv("20_Victims_of_rape.csv")   #reading the csv file
k = 573                    #Madhya Pradesh starts at index 570 in csv file
rape_cases_reported = []
rape_cases_reported2= []
j = 0 
while(k<598):
    rape_cases_reported.append(total_rapes.loc[[k],['Rape_Cases_Reported']])
    rape_cases_reported2.append(int(rape_cases_reported[j]['Rape_Cases_Reported']))
    j = j+1
    k = k+3

police_officers = pd.read_csv("12_Police_strength_actual_and_sanctioned.csv")
police_officers = police_officers[police_officers["Sub_Group_Name"] == "A2. Acual Armed Police (Incl. Women Police)"]
police_list= [] 
police_list2 = []
i = 53             #Madhya Pradesh police index starts at 53 in csv file
j = 0
while (i<334):
    police_list.append(police_officers.loc[[i],['Rank_All_Ranks_Total']])
    police_list2.append(int(police_list[j]['Rank_All_Ranks_Total']))
    i = i+35
    j = j+1

dfx = pd.DataFrame(columns=['Year','Number of police officials in Madhya Pradesh(2002-2010)','Number of Rapes(2002-2010)'])
dfx['Number of police officials in Madhya Pradesh(2002-2010)'] = police_list2
dfx['Number of Rapes(2002-2010)'] = rape_cases_reported2
Year = [2002,2003,2004,2005,2006,2007,2008,2009,2010]
dfx["Year"] = Year
sns.jointplot(x = 'Number of police officials in Madhya Pradesh(2002-2010)', y = 'Number of Rapes(2002-2010)', data = dfx, kind = 'reg')
dfx
```

# Code for Scatter plot of Kidnapping of Women vs Number of rapes in Madhya Pradesh (Graph 3)

```python
df = pd.read_csv("20_Victims_of_rape.csv")
#Cleaning unneccesary repeated values which the csv file already called, total_rape_victims
total_rape_victims = df[df.Subgroup == "Total Rape Victims"]  
#area=df.groupby('Area_Name').sum()
df_kidnapping = pd.read_csv("01_District_wise_crimes_committed_IPC_2001_2012.csv")
#df_kidnapping
df_kidnapping =df_kidnapping[df_kidnapping['STATE/UT'] == "MADHYA PRADESH"]     
df_kidnapping= df_kidnapping[df_kidnapping['YEAR'] < 2011].groupby('YEAR').sum()
df_kidnapping2 = df_kidnapping[["KIDNAPPING AND ABDUCTION OF WOMEN AND GIRLS"]]  #Extracting cases of kidnap for women
pos1 = total_rape_victims[total_rape_victims.Area_Name == "Madhya Pradesh"] #Extracting values for Madhya Pradesh
pos2 =  pos1[["Victims_of_Rape_Total"]]
#creating new dataframe
new_dataframe = pd.DataFrame(columns = ["Year","Number of Rapes in Madhya Pradesh", "KIDNAPPING AND ABDUCTION OF WOMEN AND GIRLS in Madhya Pradesh"])
a = []
b =[]
year = [2001,2002,2003,2004,2005,2006,2007,2008,2009,2010]
for j in df_kidnapping2["KIDNAPPING AND ABDUCTION OF WOMEN AND GIRLS"]:
    b.append(int(j))
for i in pos1['Victims_of_Rape_Total']:
    a.append(int(i))
new_dataframe["Number of Rapes in Madhya Pradesh"] = a
new_dataframe["KIDNAPPING AND ABDUCTION OF WOMEN AND GIRLS in Madhya Pradesh"] = b
new_dataframe["Year"] = year
sns.jointplot(x = "Number of Rapes in Madhya Pradesh", y = "KIDNAPPING AND ABDUCTION OF WOMEN AND GIRLS in Madhya Pradesh",data = new_dataframe, kind = "reg")
```

# Code for Yearly increase and decrease in the percentage of number of rapes for Madhya Pradesh and West Bengal(Graph 5)


```python
#% increase graph
rape_victims = pd.read_csv("20_Victims_of_rape.csv")
wb_rape_victims = rape_victims[rape_victims['Area_Name'] == 'West Bengal']
wb_rape_victims_by_year = wb_rape_victims.groupby("Year").sum()
plt.subplots(figsize = (15, 6))
mp_rape_victims = rape_victims[rape_victims['Area_Name'] == 'Madhya Pradesh']
mp_rape_victims_by_year = mp_rape_victims.groupby("Year").sum()
ax = (wb_rape_victims_by_year['Rape_Cases_Reported'].pct_change() * 100).plot(legend = True, 
                                                                              label = 'West Bengal')
(mp_rape_victims_by_year['Rape_Cases_Reported'].pct_change() * 100).plot(ax = ax, legend = True, 
                                                                         label = 'Madhya Pradesh')
ax.set(xlabel = 'Year', ylabel = 'Percent', 
       title = 'Yearly increase and decrease in number of rapes in West Bengal and Madhya Pradesh')
ax.axhline(0, color = 'black')
ax.axvline(2002, color = 'black')
plt.show()
print('Overall Increase in number of rapes in West Bengal =', 
      '{0:.2f}'.format(((wb_rape_victims_by_year.iloc[9]['Rape_Cases_Reported'] 
                         - wb_rape_victims_by_year.iloc[0]['Rape_Cases_Reported'])
                        /wb_rape_victims_by_year.iloc[9]['Rape_Cases_Reported']) * 100), 'Percent')
print('Overall Increase in number of rapes in Madhya Pradesh =', 
      '{0:.2f}'.format(((mp_rape_victims_by_year.iloc[9]['Rape_Cases_Reported'] 
                         - mp_rape_victims_by_year.iloc[0]['Rape_Cases_Reported'])
                        /wb_rape_victims_by_year.iloc[9]['Rape_Cases_Reported']) * 100), 'Percent')
```

# Code for horizontal bar chart of total crimes committed aginst women
